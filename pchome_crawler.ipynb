```python
brand = 'DHAF'
url_product = f'https://24h.pchome.com.tw/region/{brand}'
```

```python
driver = webdriver.Chrome()
driver.implicitly_wait(15)
driver.get(url_product)
```

```python
prdnames_elements = driver.find_elements(By.CLASS_NAME, 'prod_name') + driver.find_elements(By.CLASS_NAME, 'nick')
prdnames = list(filter(None, [element.text for element in prdnames_elements]))
print(prdnames)
print(len(prdnames))
```

```python
price_elements = driver.find_elements(By.CLASS_NAME, 'value')
price = list(filter(None, [element.text for element in price_elements]))
print(price)
print(len(price))
```

```python
slogans_elements = driver.find_elements(By.CLASS_NAME, 'slogan')
slogans = [None] * 7 + list(filter(None, [element.text for element in slogans_elements]))
print(slogans)
print(len(slogans))
```

```python
href_values = []
href_1_specific = driver.find_element(By.XPATH, '//*[@id="Block12Container7"]/div/a').get_attribute('href')
href_elements_o = driver.find_elements(By.XPATH, '//dd[@class="" and (@sys="1" or @sys="0")]')

for sys_element in href_elements_o:
    prod_img_elements = sys_element.find_elements(By.CLASS_NAME, 'prod_img')
    href_values += [img_element.get_attribute('href') for img_element in prod_img_elements]

href_values_all = [href_1_specific] + href_values

print(href_values_all)
print(len(href_values_all))
```

```python
from datetime import datetime
today_date = datetime.today().date()
df_prdnames = pd.DataFrame(prdnames, columns=['ProdNames'])
df_price = pd.DataFrame(price, columns=['Prices'])
df_slogans = pd.DataFrame(slogans, columns=['SloganTitle'])
df_href_values_all = pd.DataFrame(href_values_all, columns=['Url'])

pn_pe_sl = pd.concat([df_prdnames, df_price, df_slogans, df_href_values_all], axis=1)
pn_pe_sl['date'] = today_date

# 顯示 DataFrame
print(pn_pe_sl)
```

```python
target_a_elements = driver.find_elements(By.CSS_SELECTOR, '[data-testid="post-title"][aria-label]')
if target_a_elements:
    aria_label_values = [a.get_attribute('aria-label') for a in target_a_elements]
df_title = pd.DataFrame({'Title': aria_label_values})
```

```python
print(df_title)
```

```python
time_parent_elements = driver.find_elements(By.CSS_SELECTOR, '[datetime]')

post_time_list = []
how_far_list = []

for element in time_parent_elements:
    datetime_str = element.get_attribute('datetime')
    parsed_datetime = datetime.strptime(datetime_str, "%Y-%m-%dT%H:%M:%S.%fZ")
    date_part = parsed_datetime.strftime("%Y-%m-%d")
    how_far = element.text

    post_time_list.append(date_part)
    how_far_list.append(how_far)

df_post = pd.DataFrame({
    'Post-Time': post_time_list,
    'How-Far': how_far_list
})
```

```python
print(df_post)
```

```python
text_content_elements = driver.find_elements(By.CSS_SELECTOR, '.text-neutral-content-weak.text-12')

vote_list = []
comments_list = []

for text_content in text_content_elements:
    for span_element in text_content.find_elements(By.TAG_NAME, 'span'):
        if 'votes' in span_element.text or 'vote' in span_element.text:
            votes_elements = span_element.find_elements(By.TAG_NAME, 'faceplate-number')
            vote = votes_elements[0].text if votes_elements else "N/A"
            vote_list.append(vote)
        elif 'comments' in span_element.text or 'comment' in span_element.text:
            comments_elements = span_element.find_elements(By.TAG_NAME, 'faceplate-number')
            comment = comments_elements[0].text if comments_elements else "N/A"
            if comment != "N/A":
                comments_list.append(comment)

df_vote_comments = pd.DataFrame({
    'Vote': vote_list,
    'Comments': comments_list
})
```

```python
df_all = pd.merge(pd.merge(df_title, df_post, left_index=True, right_index=True), df_vote_comments, left_index=True, right_index=True)
```

```python
df_all.to_excel('output3_file.xlsx', index=False)
```
